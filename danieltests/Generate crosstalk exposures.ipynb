```python
import numpy as np
```

```python
minexp=1.0 #Value for lowest exposure time
minlamp=1  #lowest lamp value
topexp=3 #an exposure then when multiplied by toplamp gives you the max desired adu
toplamp=66.67

maxexp=15.0
maxlamp=67.9
bad1=6.4
bad2=12
resolution=100
```

```python
ints=np.linspace(minlamp*minexp,topexp*toplamp,resolution)
exp=minexp
for i in ints:
    if minlamp<=i/exp<=bad1 or bad2<=i/exp<=maxlamp:
        print(str(exp)+"   "+str("{:.2f}".format(i/exp))+",")
    elif bad1<=i/exp<=bad2:
        print(str(exp*2)+"   "+str("{:.2f}".format(i/(exp*2)))+",")
    elif i/exp>=maxlamp:
        while i/exp>=maxlamp:
            exp+=0.5
        print(str(exp)+"   "+str("{:.2f}".format(i/exp))+",")
```

```python
ints=np.linspace(minlamp*minexp,topexp*toplamp,resolution)
exp=minexp
for i in ints:
    if minlamp<=i/exp<=bad1 or bad2<=i/exp<=maxlamp:
        print(str(exp)+"   "+str("{:.2f}".format(i/exp))+"  2,")
    elif bad1<=i/exp<=bad2:
        print(str(exp*2)+"   "+str("{:.2f}".format(i/(exp*2)))+"  2,")
    elif i/exp>=maxlamp:
        while i/exp>=maxlamp:
            exp+=0.5
        print(str(exp)+"   "+str("{:.2f}".format(i/exp))+"  2,")
```

```python
ints=np.linspace(minlamp*minexp,topexp*toplamp,resolution)
exp=minexp
listexp=[]
listlamp=[]
for i in ints:
    if minlamp<=i/exp<=bad1 or bad2<=i/exp<=maxlamp:
        listexp.append(str(exp))
        listlamp.append(str("{:.2f}".format(i/exp)))
        #print(str(exp)+"   "+str("{:.2f}".format(i/exp))+",")
    elif bad1<=i/exp<=bad2:
        #print(str(exp*2)+"   "+str("{:.2f}".format(i/(exp*2)))+",")
        listexp.append(str(exp*2))
        listlamp.append(str("{:.2f}".format(i/(exp*2))))
    elif i/exp>=maxlamp:
        while i/exp>=maxlamp:
            exp+=0.5
        listexp.append(str(exp))
        listlamp.append(str("{:.2f}".format(i/exp)))
        #print(str(exp)+"   "+str("{:.2f}".format(i/exp))+",")
for j in range(len(listexp)):
    print(listexp[j]+"   "+listlamp[j]+",")
for k in range(len(listexp)):
    print(listexp[-k-1]+"   "+listlamp[-k-1]+",")
```
