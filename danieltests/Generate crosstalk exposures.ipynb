```python
import numpy as np
```

```python
minexp=1.0 #Value for lowest exposure time
minlamp=0.3  #lowest lamp value
topexp=2 #an exposure then when multiplied by toplamp gives you the max desired adu
toplamp=60

maxexp=15.0
maxlamp=67.9
bad1=6.4
bad2=12
resolution=100
```

```python
#standard
ints=np.linspace(minlamp*minexp,topexp*toplamp,resolution)
exp=minexp
for i in ints:
    if minlamp<=i/exp<=bad1 or bad2<=i/exp<=maxlamp:
        print(str(exp)+"   "+str("{:.2f}".format(i/exp))+",")
    elif bad1<=i/exp<=bad2:
        print(str(exp*2)+"   "+str("{:.2f}".format(i/(exp*2)))+",")
    elif i/exp>=maxlamp:
        while i/exp>=maxlamp:
            exp+=0.5
        print(str(exp)+"   "+str("{:.2f}".format(i/exp))+",")
```

```python
#flat pairs
ints=np.linspace(minlamp*minexp,topexp*toplamp,resolution)
exp=minexp
for i in ints:
    if minlamp<=i/exp<=bad1 or bad2<=i/exp<=maxlamp:
        print(str(exp)+"   "+str("{:.2f}".format(i/exp))+"  2,")
    elif bad1<=i/exp<=bad2:
        print(str(exp*2)+"   "+str("{:.2f}".format(i/(exp*2)))+"  2,")
    elif i/exp>=maxlamp:
        while i/exp>=maxlamp:
            exp+=0.5
        print(str(exp)+"   "+str("{:.2f}".format(i/exp))+"  2,")
```

```python
#hysteresis
ints=np.linspace(minlamp*minexp,topexp*toplamp,resolution)
exp=minexp
listexp=[]
listlamp=[]
for i in ints:
    if minlamp<=i/exp<=bad1 or bad2<=i/exp<=maxlamp:
        listexp.append(str(exp))
        listlamp.append(str("{:.2f}".format(i/exp)))
        #print(str(exp)+"   "+str("{:.2f}".format(i/exp))+",")
    elif bad1<=i/exp<=bad2:
        #print(str(exp*2)+"   "+str("{:.2f}".format(i/(exp*2)))+",")
        listexp.append(str(exp*2))
        listlamp.append(str("{:.2f}".format(i/(exp*2))))
    elif i/exp>=maxlamp:
        while i/exp>=maxlamp:
            exp+=0.5
        listexp.append(str(exp))
        listlamp.append(str("{:.2f}".format(i/exp)))
        #print(str(exp)+"   "+str("{:.2f}".format(i/exp))+",")
for j in range(len(listexp)):
    print(listexp[j]+"   "+listlamp[j]+",")
for k in range(len(listexp)):
    print(listexp[-k-1]+"   "+listlamp[-k-1]+",")
```

```python
#flat pairs hysteresis
ints=np.linspace(minlamp*minexp,topexp*toplamp,resolution)
exp=minexp
for i in ints:
    if minlamp<=i/exp<=bad1 or bad2<=i/exp<=maxlamp:
        print(str(exp)+"   "+str("{:.2f}".format(i/exp))+"  2,")
    elif bad1<=i/exp<=bad2:
        print(str(exp*2)+"   "+str("{:.2f}".format(i/(exp*2)))+"  2,")
    elif i/exp>=maxlamp:
        while i/exp>=maxlamp:
            exp+=0.5
        print(str(exp)+"   "+str("{:.2f}".format(i/exp))+"  2,")
for j in range(len(ints)):
    if minlamp<=ints[-j-1]/exp<=bad1 or bad2<=ints[-j-1]/exp<=maxlamp:
        print(str(exp)+"   "+str("{:.2f}".format(ints[-j-1]/exp))+"  2,")
    elif bad1<=ints[-j-1]/exp<=bad2:
        print(str(exp*2)+"   "+str("{:.2f}".format(ints[-j-1]/(exp*2)))+"  2,")
    elif ints[-j-1]/exp>=maxlamp:
        while ints[-j-1]/exp>=maxlamp:
            exp+=0.5
        print(str(exp)+"   "+str("{:.2f}".format(ints[-j-1]/exp))+"  2,")
```

```python
#flat pairs random order
ints=np.linspace(minlamp*minexp,topexp*toplamp,resolution)
exp=minexp
for j in range(resolution):
    index=np.random.randint(0,len(ints))
    i=ints[index]
    if minlamp<=i/exp<=bad1 or bad2<=i/exp<=maxlamp:
        print(str(exp)+"   "+str("{:.2f}".format(i/exp))+"  2,")
    elif bad1<=i/exp<=bad2:
        print(str(exp*2)+"   "+str("{:.2f}".format(i/(exp*2)))+"  2,")
    elif i/exp>=maxlamp:
        while i/exp>=maxlamp:
            exp+=0.5
        print(str(exp)+"   "+str("{:.2f}".format(i/exp))+"  2,")
    ints=np.delete(ints,index)
```
